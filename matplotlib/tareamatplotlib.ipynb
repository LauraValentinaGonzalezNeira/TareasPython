```python
import matplotlib.pyplot as plt
unidades=[4976 , 4713 , 3890 , 3705 , 3653 , 3469]
caballos=[186,66,160,111,143,156]
precio=[108,44,52.7,50,74,48.8]
vehiculo=['Mazda cx 30','Renault Kwind','Chevrolet','Renault Stepway','Renault Duster','Renault Sandero']
colores=['red','orange','blue','green','white','magenta']
plt.pie(unidades,labels=vehiculo,colors=colores,wedgeprops={'edgecolor':'black'})
plt.title('Venta de los carros')
plt.show()
```

```python
vehiculo1=['Mazda','Renault K','Chevrolet','Renault St','Renault D','Renault San']
plt.plot(unidades,precio,color='yellow',linewidth=3,label='Precio')
plt.plot(unidades,caballos,color='green',linewidth=3,label='Caballos de fuerza')
plt.legend()
plt.title('Precio y Caballos de Fuerza vs Unidades Vendidas')
plt.xlabel('Unidades vendidas')
plt.ylabel('Precio - Caballos de Fuerza')
plt.show()
```
